### SOURCE MEMORY TEST

```python
newdf2
## correct context

tst_obj = []
cor_cntxt = []
tst_val = []
all_safe = []
all_threat = []

for index, row in newdf2.iterrows():
    tst_obj.append(newdf2.objects_pics[index])
    cor_cntxt.append(newdf2.contexts_pics[index])
    tst_val.append(newdf2.context_type[index])
    if newdf2.context_type[index] == 'SAFE':
        all_safe.append(newdf2.contexts_pics[index])
    if newdf2.context_type[index] == 'THREAT':
        all_threat.append(newdf2.contexts_pics[index])
```

```python
## foil contexts
same_foil = []
same_val = []
opp_foil = []
opp_val = []
for index, row in newdf2.iterrows():
    tst = newdf2.contexts_pics[index]
    if tst in all_safe:
        b = [x for x in all_safe if x != tst]
        foil = choice(b)
        foil2 = choice(all_threat)
        same_foil.append(foil)
        opp_foil.append(foil2)
        same_val.append('safe')
        opp_val.append('threat')
    if tst in all_threat:
        b = [x for x in all_threat if x != tst]
        foil = choice(b)
        foil2 = choice(all_safe)
        same_foil.append(foil)
        opp_foil.append(foil2)
        same_val.append('threat')
        opp_val.append('safe')

srcdf = pd.DataFrame(
    {'test_object': tst_obj,
     'correct_context': cor_cntxt,
    'tested_valence': tst_val,
    'same_val': same_val,
    'opp_val': opp_val,
    'same_foil': same_foil,
    'opp_foil': opp_foil})     
```
